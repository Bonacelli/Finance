# NumPy - Operações

```python
import numpy as np
```

## Diferenças entre Listas Python a Numpy Arrays

### **Vetorização**

```python
lista = [0,1,2,3,4,5]
```

```python
lista
```

```python
vetor = np.array(lista)
```

```python
vetor
```

```python
lista * 2
```

```python
vetor * 2 
```

```python
# Gera um erro
lista + 1
```

```python
vetor + 1 
```

### **Broadcasting**

broad.png

```python
A = np.array([[0,0,0],[10,10,10], [20,20,20], [30,30,30]])
B = np.array([[0,1,2],[0,1,2],[0,1,2],[0,1,2]])
```

```python
A
```

```python
B
```

```python
A + B
```

```python
A + [0,1,2]
```

```python
np.array([[0],[10],[20],[30]]) + [0,1,2]
```
